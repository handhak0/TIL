# 01. Catagorical 

- one of a limited number of possible values (정해진 값들 중에서만 선택 가능) 
- symbols (더하기 빼기 등의 의미가 없음) 
- order relationships (순서에는 의미가 있을 수 있음) 

- Nominal(단순 이름), Ordinal(순서) 

```python
import pandas as pd 
```

```python
s = pd.Series(["손","김","이","양"])
s # 데이터 타입이 object임 
```

```python
ctg = s.astype("category")
ctg # 데이터 타입을 category로 변환시킴 
```

```python
# cat 속성 
cat_s = pd.Series(["a","b","c","d","a","b","c","d"]).astype("category")
cat_s
```

```python
cat_s.cat.categories # 카테고리의 종류를 확인 
```

```python
cat_s.cat.codes # 카테고리의 인덱스 확인 
```

```python
# 카테고리 추가
actual = ['a','b','c','d','e']

cat_s2 = cat_s.cat.set_categories(actual)
cat_s2 # e가 없지만, 카테고리의 종류에 e가 추가됨
```

```python
# 사용 안하는 카테고리 삭제 
cat_s2.cat.remove_unused_categories()
```
