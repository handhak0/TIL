#00. General Characteristics of DataSets 
- Dimensionality 
   - 데이터에 포함된 attribute 개수 
   - 차원이 높으면 분석이 좋지 않기에 차원축소 필요 
- Distribution 
   - attribute 별 분포 
   - 분포가 정규분포면 표현하기 쉬움 
   - Skewness는 분석 어렵게 함
